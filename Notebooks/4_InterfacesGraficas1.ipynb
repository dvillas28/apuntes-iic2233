# Eventos y Se√±ales

La estructura de funcionamiento de una **gui basada en manejo de eventos**

El loop de eventos se inicia con el metodo `exec()` de `QApplication`

## Elementos Fundamentales: 

1. **Fuente del evento**:  Objeto que genera el cambio de estado, delega tareas
2. **Objeto evento**: Objeto que *encapsula el cambio de evento*
3. **Objeto Destino**: Objeto al cual se le desea avisar del cambio de estado

- La **Fuente del evento** maneja el evento pasandole el **objeto destino** al **objeto evento**

## ***signal*** y ***slot***

- mecanismo que pertime el manejo de eventos
- cuando un evento ocurre, el objeto emite una *signal* que a un *slot* respectivo, donde este es un
elemento *callable* de Python (funcion o metodo)

## `sender()` el emisor de la *signal*

- obtiene una referencia al objeto que recibio la se√±al
- como se supone que conocemos al sender, podemos acceder a sus metodos para
poder manipularlo
- tambien podemos manipular otros objetos desde el slot 

## Eventos de Mouse

- todo widgets tiene distintos eventos que se pueden conectar distintos slots

- `mousePressEvent()`: metodo incluido en todos los objetos clase `QWidget`, se hace cargo del comportamiento del widget al presionar al mouse sobre el. Respresenta el evento generico de clic sobre un widget
- No es necesario realizar la conxion evento-metodo con este tipo de eventos, la conexion ya se encuentra definida y solo se debe sobreescribir el metodo
- sobreescribir `mousePressEvent` significa definir el codigo que se ejecuta cuando se clickea ese Widget

- `mouseReleaseEvent()` y `mouseMoveEvent` funcionan de la misma manera

## Eventos de teclado

- `keyPressEvent`. por defecto no hace nada, le hecemos override para otorgarle a una funcion
