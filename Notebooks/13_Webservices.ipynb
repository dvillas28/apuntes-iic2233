# 1. Webservices

Conjunto de aplicaciones **cliente-servidor** comunicadas por la web

## API (*Application Programming Interface*)

- Conjunto de funciones expuestas por el servidor para poder ser utilizadas por el cliente u otros programas.
- Metodos expuestos por un sitio web, muchos servicios web exponen una API para que otras aplicaciones se conecten a ellas.

## URL (*Uniform Resource Locator*)

- texto que nos ayuda a localizar algo en internet
- Formato para poder apuntar correctamente a los que buscamos

## HTTP (*Hypertext Transfer Protocol*)

- Protocolo de transacciones de comuniacion entre clientes y servidores
- Funciona como protocolo **request-response**. El cliente realiza una peticion y el server responde con la info solicitada.

- El servidor no guarda info sobre el cliente. Para acceder a un metodo que requiere identificacion, se debe enviar algo que acredite la identidad

- El funcionamiento del servidor esta basado en metodos que indican la accion a desarrollar

- Y el servidor posse una serie de **codigos de estado**, usados para entregar info al cliente acerca del resultado de la peticion
    - `200`. OK, solicitud exitosa
    - `403`. Solicitud correcta, pero se rechaza dar respuesta
    - `404`. Recurso solicitado no se encuentra en el servidor
    - `500`. Error interno del servidor

- El *Body* es el contenido actual de la request, dependiendo del server (de la API), puede ser un archivo de texto conteniendo HTML, un json, etc

### *headers*
- Los *headers* incluyen informacion interna para autenticaciones o encoding de la info
- Podemos incluir un **token especial** de acceso que a la API le sirve para identificar y verificar los permisos


## Script del lado del cliente - `requests`

```python
import requests

# solicitud GET
url = "https://www.google.cl"
response = requests.get(url)
```

```python
# vemos el resultado de nuestra peticion
print(response.status_code) # 200: todo OK
```

```python
# vemos el contenido
print(response.text) # doc html de google para cargar
```

### Ejemplo Github

```python
response = requests.get('https://api.github.com/repos/IIC2233/Syllabus/issues/143')
```

```python
# print(response.json())

for i in response.json():
    print(f'{i}: {response.json()[i]}')
```

```python
# imprimimos el cuerpo solamente
print(response.json()["body"])
```
